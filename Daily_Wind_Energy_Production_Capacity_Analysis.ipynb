# In this project I've done some analysis and visualization on Daily Wind Energy Production Capacity of australia with the data available at [Aneroid Energy](https://anero.id/energy/wind-energy).

**In this project I've built a web scraper to scrape the data from the public domain and made some analysis & visualization with the scraped data.**

**This project gives the audience a live visual on the data from the website whenever it's updated all you need to do is just run each and every cell to get the new visual on the available data in an instant.**

***This project is a work in progress but all the visual given in this workspace is constantly checked and updated inorder to give the accurate visual analysis.***

```python
#Imported the necessary packages and libraries
import matplotlib.pyplot as plt
from bs4 import BeautifulSoup
import plotly.express as px
import pandas as pd
import requests
import csv

#This is the web scraper
request = requests.get('https://anero.id/energy/wind-energy')

soup = BeautifulSoup(request.text, 'html5lib')

wind_t = soup.select_one('table')

table_body = wind_t.select_one('tbody')

south_whales = table_body.select('tr')[1:17]

#print(south_whales)

queensland = table_body.select('tr')[19:21]

south_australia = table_body.select('tr')[23:45]

tasmania = table_body.select('tr')[47:51]

victoria = table_body.select('tr')[53:78]

csv_data = [['farm_Name', 'capacity']]

#This function is used to write the collected data into a CSV file 
def csv_writer():
  with open('Australia Dataset.csv', 'w') as file:
    write = csv.writer(file)
    write.writerows(csv_data)

def data_collection(country):
  try:
    for e in country:
      name = e.find_all('td')[1].text
      capacity = e.find_all('td')[3].text
      csv_data.append([name, capacity])
    csv_writer()
  except:
    print("Something went wrong")

#This function is for reading the generated csv file
def csv_reader(path, filename):
  return pd.read_csv(path+'/'+filename)

path = '/content'

#Run all the functions to get the energy capacity of austalia as a whole
#To get state wise data for the energy capacity of austalia run the below functions one by one and change the name of the csv for each dataset
#While running the below functions one by one comment out the rest of the function to get the dataset of a specific country

data_collection(south_whales)

data_collection(queensland)

data_collection(south_australia)

data_collection(tasmania)

data_collection(victoria)

aus_wind = csv_reader(path, 'Australia Dataset.csv')

sw = csv_reader(path, 'New South Whales Dataset.csv')

q =  csv_reader(path, 'Queensland Dataset.csv')

sa = csv_reader(path, 'South Australia Dataset.csv')

t = csv_reader(path, 'Tasmania Dataset.csv')

v = csv_reader(path, 'Victoria Dataset.csv')
```
